## Training ANN

### Step 1 | Import libraries

```python
import numpy as np
import pandas as pd
import keras
import tensorflow
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report
from plot_confusion_matrix import plot_confusion_matrix
from keras.callbacks import EarlyStopping
import tqdm
```

### Step 2 | Import preprocessed dataset

```python
train_df = pd.read_csv('./data/preprocessed_train_df.csv')
test_df = pd.read_csv('./data/preprocessed_test_data.csv')
```

```python
X_train = train_df.drop(columns=['NObeyesdad'])
y_train = train_df['NObeyesdad']
X_test = test_df.drop(columns=['NObeyesdad'])
y_test = test_df['NObeyesdad']
```

```python
# Feature scaling the dataset by StandardScaler (Min-Max Scaling)

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

### Step 3 | Building a Neural Network

#### Step 3.1 | Define a neural network

Our neural network is a multi-layer perceptron (MLP) designed for a multi-class classification task with 7 output classes. It consists of the following:

- Input Layer: Takes input vectors of 20 features.

- Hidden Layers: Three hidden layers, each with 64 neurons and tanh activation functions, designed to learn complex representations of the input data.

- Dropout Layers: Three dropout layers, each with a 10% dropout rate, included after each hidden layer to prevent overfitting by randomly dropping 10% of the neurons during training.

- Output Layer: A dense layer with 7 neurons and a softmax activation function to output the probabilities of the 7 classes.

The model is compiled with the Adam optimizer and uses sparse categorical cross-entropy as the loss function, suitable for multi-class classification with integer labels. The primary performance metric tracked during training is accuracy.

```python
classifier = Sequential()
classifier.add(Dense(units = 64, activation = 'tanh', input_dim = 20))
classifier.add(Dropout(0.1))
classifier.add(Dense(units = 64, activation = 'tanh'))
classifier.add(Dropout(0.1))
classifier.add(Dense(units = 64, activation = 'tanh'))
classifier.add(Dropout(0.1))
classifier.add(Dense(units = 7, activation = 'softmax'))
classifier.compile(optimizer = 'adam', loss = 'sparse_categorical_crossentropy', metrics = ['accuracy'])
```

The `EarlyStopping` callback is employed to enhance the efficiency and performance of our neural network training. 

- It monitors the validation loss (`val_loss`) and halts training if no improvement is observed for 8 consecutive epochs (`patience=8`). This approach helps in preventing overfitting by stopping the training process before the model starts to learn noise from the training data. 

- `restore_best_weights=True`, it ensures that the model retains the best weights from the epoch with the lowest validation loss, providing the optimal version of the model for deployment. Overall, `EarlyStopping` saves computational resources and time while improving the generalization of the model.

```python
early_stopping = EarlyStopping(monitor='val_loss', patience=7, restore_best_weights=True)
```

```python
classifier.summary()
```

#### Step 3.2 | Training ANN

```python
history = classifier.fit(X_train, y_train, epochs=50, batch_size=32, validation_split=.2, callbacks=[early_stopping])
```

### Step 4 | Evaluation

```python
y_pred = np.argmax(classifier.predict(X_test), axis=1)
```

```python
print(classification_report(y_test, y_pred, digits = 4))
```

```python
plot_confusion_matrix(y_test, y_pred, "Confusion matrix of ANN")
```

```python
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training and Validation Loss')
plt.legend()
plt.show()
```

```python
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training and Validation Accuracy')
plt.legend()
plt.show()
```
